```python
import json

def clean_data(data):
    # 1. Remove users with empty or whitespace-only names
    data["users"] = [user for user in data["users"] if user["name"].strip()]
    
    # 2. Remove duplicate friend IDs (e.g., [2, 2] -> [2])
    for user in data["users"]:
        user["friends"] = list(set(user["friends"]))

    data['users'] = [
        user for user in data['users'] 
        if user['name'].strip() and (user['friends'] or user['liked_pages'])
    ]
                    
        
    # 3. Handle duplicate Page IDs (keeping the first occurrence)
    unique_pages = []
    seen_page_ids = set()
    for page in data["pages"]:
        if page["id"] not in seen_page_ids:
            unique_pages.append(page)
            seen_page_ids.add(page["id"])
    data["pages"] = unique_pages
                                         
    return data

# Load, clean, and save
try:
    with open("data2.json", "r") as f:
        data = json.load(f)

    data = clean_data(data)

    with open("cleaned_data2.json", "w") as f:
        json.dump(data, f, indent=4)

    print("Data has been cleaned and saved to cleaned_data2.json")
except FileNotFoundError:
    print("Error: data2.json not found in the pythonrefresher directory.")
```
