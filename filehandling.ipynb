```python
import sys
import numpy as np

# Create a list with 1000 elements
list_data = list(range(1000))

# Create a numpy array from the same list
numpy_data = np.array(list_data)

# Calculate and print the memory size of the Python list
# Note: sys.getsizeof(list_data) returns the size of the list object itself
print("Python list size:", sys.getsizeof(list_data) * len(list_data), "bytes")

# Calculate and print the memory size of the NumPy array
# nbytes returns the total bytes consumed by the elements of the array
print("NumPy array size:", numpy_data.nbytes, "bytes")
```
