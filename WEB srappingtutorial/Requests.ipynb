```python
soup = BeautifulSoup(a.text, "html.parser")
```

```python
h3s = soup.find_all("h3")
```

```python
articles = soup.select("article.product_pod")
```

```python
items = []
for article in articles:
    title = article.h3.a["title"]
    price = article.select("p.price_color")[0].text.split("£")[1]
    rating_element = article.select("p.star-rating")[0]
    rating_classes = rating_element.get("class", [1])
    items.append((title, price, rating_classes))
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(items, columns=["Title", "Price", "Rating"])
print(df.head())
```
