```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1, 2, 3], [4, 5, 1])
plt.show()
```

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
runs = [1000, 1500, 2000, 2500, 3000, 3500, 4000]

plt.plot(years, runs)
plt.show()
```

```python
plt.plot(years, runs)
plt.xlabel("Years")
plt.ylabel("Runs")
plt.title("Runs scored by Sachin in ODIs")
plt.show()
```

```python
kohli_runs = [100, 200, 300, 400, 500, 600, 700]
sachin_runs = [1000, 1500, 2000, 2500, 3000, 3500, 4000]    
plt.plot(years, sachin_runs, label="Sachin")
plt.plot(years, kohli_runs, label="Kohli")
plt.xlabel("Years")
plt.ylabel("Runs")
plt.title("Runs scored by Sachin and Kohli in ODIs")
plt.legend()
plt.show()
```

```python
plt.plot(years, kohli_runs, "ro--", label="Kohli")
plt.plot(years, sachin_runs, "go--", label="Sachin")
plt.legend()
```

```python
plt.plot(years, kohli_runs, color = "red", label="Kohli")
plt.plot(years, sachin_runs, color = "green", label="Sachin")
plt.legend()
```

```python
plt.tight_layout()
```

```python
import numpy as np
kohli_runs = [100, 200, 300, 400, 500, 600, 700]
sachin_runs = [1000, 1500, 2000, 2500, 3000, 3500, 4000]    
sehwag_runs = [200, 300, 400, 500, 600, 700, 800]

x = np.arange(len(years))
width = 0.25
```

```python
x
```

```python
plt.bar(x - width, kohli_runs, width=width, label="Kohli")
plt.bar(x, sachin_runs, width=width, label="Sachin")
plt.bar(x + width, sehwag_runs, width=width, label="Sehwag")

plt.xlabel("Years")
plt.ylabel("Runs")
plt.title("Runs scored by Sachin, Kohli and Sehwag in ODIs")
plt.xticks(x, years)    
plt.legend()
plt.tight_layout()
plt.show()
```

```python
players = ["Kohli", "Sachin", "Sehwag"]

runs_yrs = [100+200+300+400+500+600+700, 1000+1500+2000+2500+3000+3500+4000, 200+300+400+500+600+700+800    ]
plt.barh(players, runs_yrs)
plt.xlabel("Players")   
plt.ylabel("Runs")
plt.title("Total runs scored by Sachin, Kohli and Sehwag in ODIs    ")
plt.show()
```
