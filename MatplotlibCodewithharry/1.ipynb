```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
runs = [1000, 1500, 2000, 2500, 3000, 3500, 4000]

plt.plot(years, runs)
plt.show()
```

```python
plt.plot(years, runs)
plt.xlabel("Years")
plt.ylabel("Runs")
plt.title("Runs scored by Sachin in ODIs")
plt.show()
```

```python
kohli_runs = [100, 200, 300, 400, 500, 600, 700]
sachin_runs = [1000, 1500, 2000, 2500, 3000, 3500, 4000]    
plt.plot(years, sachin_runs, label="Sachin")
plt.plot(years, kohli_runs, label="Kohli")
plt.xlabel("Years")
plt.ylabel("Runs")
plt.title("Runs scored by Sachin and Kohli in ODIs")
plt.legend()
plt.show()
```

```python
plt.plot(years, kohli_runs, "ro--", label="Kohli")
plt.plot(years, sachin_runs, "go--", label="Sachin")
plt.legend()
```

```python
plt.plot(years, kohli_runs, color = "red", label="Kohli")
plt.plot(years, sachin_runs, color = "green", label="Sachin")
plt.legend()
```

```python
import numpy as np
kohli_runs = [100, 200, 300, 400, 500, 600, 700]
sachin_runs = [1000, 1500, 2000, 2500, 3000, 3500, 4000]    
sehwag_runs = [200, 300, 400, 500, 600, 700, 800]

x = np.arange(len(years))
width = 0.25
```

```python
x
```

```python
plt.bar(x - width, kohli_runs, width=width, label="Kohli")
plt.bar(x, sachin_runs, width=width, label="Sachin")
plt.bar(x + width, sehwag_runs, width=width, label="Sehwag")

plt.xlabel("Years")
plt.ylabel("Runs")
plt.title("Runs scored by Sachin, Kohli and Sehwag in ODIs")
plt.xticks(x, years)    
plt.legend()
plt.tight_layout()
plt.show()
```

```python
players = ["Kohli", "Sachin", "Sehwag"]

runs_yrs = [100+200+300+400+500+600+700, 1000+1500+2000+2500+3000+3500+4000, 200+300+400+500+600+700+800    ]
plt.barh(players, runs_yrs)
plt.xlabel("Players")   
plt.ylabel("Runs")
plt.title("Total runs scored by Sachin, Kohli and Sehwag in ODIs    ")
plt.show()
```

```python
plt.style.use("ggplot")
labels = ["Kohli", "Sachin", "Sehwag"]
runs = [100+200+300+400+500+600+700, 1000+1500+2000+2500+3000+3500+4000, 200+300+400+500+600+700+800    ]
plt.pie(runs, labels=labels, autopct="%0.2f%%") 
plt.title("Total runs scored by Sachin, Kohli and Sehwag in ODIs    ")  
plt.show()
```

```python
plt.pie(
    runs,
    labels=labels,
    colors=["red", "green", "blue"],
    wedgeprops={"edgecolor": "black", "fill": True, "linewidth": 2},
    autopct="%0.2f%%"

)

plt.show()
```

```python
# runs/labels length is 3,
# ode must also be length 3
explore = [0, 0.1, 0]
plt.pie(
    runs,
    labels=labels,
    colors=["red", "green", "blue"],
    explode=explore,
    shadow=True,
    autopct="%0.2f%%"
    
)
plt.title("Total runs scored by Sachin, Kohli and Sehwag in ODIs    ")
plt.show()
```

```python
activities = ["Sleeping", "Eating", "Working", "Playing"]
```

```python
time_spent = [8, 2, 8, 6]
colors = ["red", "green", "blue", "cyan"]
plt.figure(figsize=(10, 5))
plt.pie(time_spent, labels=activities, colors=colors, autopct="%0.2f%%", startangle=90) 
plt.title("My daily activities")    
plt.axis("equal")
```

```python
days = [1, 2, 3, 4, 5, 6, 7]
studying = [2, 3, 4, 5, 6, 7, 8]
playing = [1, 2, 3, 4, 5, 6, 7]
watching_tv = [1, 2, 3, 4, 5, 6, 7]
sleeping = [8, 8, 8, 8, 8, 8, 8]

labels = ["Studying", "Playing", "Watching TV", "Sleeping"]
colors = ["red", "green", "blue", "cyan"]
plt.figure(figsize=(10, 5))
plt.stackplot(days, studying, playing, watching_tv, sleeping, labels=labels, colors=colors)
plt.legend(loc="upper left")
plt.grid()
```
